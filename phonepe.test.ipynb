```python
#aggre_transcation

path1="C:/Users/kumar/OneDrive/Desktop/phone_pe/pulse/data/aggregated/transaction/country/india/state/"
aggre_trans_list=os.listdir(path1)

columns1={"states":[],"years":[],"Quarter":[],"Transcation_type":[],"Transcation_count":[],"Transcation_amount":[] }

for state in aggre_trans_list:
    cur_states=path1+state+"/"
    aggre_year_list=os.listdir(cur_states)

    for year in aggre_year_list:
        cur_year=cur_states+year+"/" 
        aggre_file_list=os.listdir(cur_year)
        
        for file in aggre_file_list:
            cur_file=cur_year+file
            data=open(cur_file,"r")

            A=json.load(data)

            for i in A["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                columns1["Transcation_type"].append(name)
                columns1["Transcation_count"].append(count)
                columns1["Transcation_amount"].append(amount)
                columns1["states"].append(state)
                columns1["years"].append(year)
                columns1["Quarter"].append(int(file.strip(".json")))
                
aggre_transcation=pd.DataFrame(columns1)

aggre_transcation["states"]=aggre_transcation["states"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
aggre_transcation["states"]=aggre_transcation["states"].str.replace('-',' ')
aggre_transcation["states"]=aggre_transcation["states"].str.title()
aggre_transcation["states"]=aggre_transcation["states"].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```
